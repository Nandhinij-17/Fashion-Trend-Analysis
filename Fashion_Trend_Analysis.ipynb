```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import plotly.express as px

df=pd.read_excel("fashion_data.xls")

print(df.head())
print(df.info())
print(df.describe())
print(df.tail())
df.drop(2, inplace=True)
print(df.head())
print(df.isnull())

print(df.groupby("gender")["price"].sum())
print(df.groupby("gender")["price"].max())
print(df.groupby("season")["price"].mean())
print(df.groupby("category")["price"].mean())
print(df.groupby("category")["price"].median())

print(df["category"].value_counts())
print(df.sort_values("season"))
print(df.info())

plt.plot(df["gender"],df["price"])
plt.show()

plt.title("Fashion Chart")
plt.xlabel("gender")
plt.ylabel("price")
plt.grid(True)
plt.show()

plt.bar(df["product_name"],df["reviews_count"])
plt.title("Bar Chart")
plt.show()

plt.pie(df["sales_count"],
        labels=df["color"],
        autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()

plt.hist(df["price"],bins=5,alpha=0.5, label="price",color="pink",edgecolor="black")
plt.figure(figsize=(8,5))
sns.countplot(x="category", data=df)
plt.title("Category Count")
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(7,4))
sns.scatterplot(x="price", y="reviews_count", hue="material", data=df)
plt.title("Scatter Plot")
plt.xlabel("price")
plt.ylabel("reviews count")
plt.show()

plt.figure(figsize=(7,4))
sns.boxplot(x="sales_count", y="average_rating", hue="material", data=df)
plt.title("Scatter Plot")
plt.xlabel("sales_count")
plt.ylabel("average_rating")
plt.show()

plt.figure(figsize=(7,4))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

plt.figure(figsize=(7,4))
sns.pairplot(df)
plt.title("Pair Plot")
plt.show()

plt.figure(figsize=(10,5))
plt.plot(df["product_name"], df["sales_count"],
         marker="o",
         linestyle="-",
         color="green")
plt.title("Product Sales Count")
plt.xlabel("Product Name")
plt.ylabel("Sales Count")
plt.grid(True)
plt.show()

plt.figure(figsize=(7,5))
sns.violinplot(x="gender", y="price", data=df)
plt.title("Price Distribution")
plt.show()

plt.figure(figsize=(7,5))
sns.boxplot(x="gender", y="price", data=df)
plt.title("Price Distribution by Gender")
plt.show()

fig = px.scatter(
    df,
    x="price",
    y="sales_count",
    color="category",
    hover_name="product_name",
    title="Price vs Sales Count"
)
fig.show()

plt.figure(figsize=(7,5))
sns.regplot(x="price", y="sales_count", data=df)
plt.title("Price vs Sales Count Regression")
plt.show()
```
